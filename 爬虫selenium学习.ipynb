```python

# -*- coding:utf-8 -*-
from selenium import webdriver
import time
 
def noWindow():
    # 不开启浏览器得到返回值
    # 加载启动项，只开一个监听接口获取返回值
    option = webdriver.ChromeOptions()
    option.headless = True  # 不打开浏览器，后台运行
    
    # 通过url访问浏览器
    driver = webdriver.Chrome(chrome_options=option)
    url = "https://www.baidu.com"

    driver.get(url)
    
    time.sleep(3)
    driver.quit()

def MaxWindow():
    # 最大化浏览器
    driver = webdriver.Chrome()
    url = "https://www.baidu.com"
    driver.get(url)
    driver.maximize_window()
    
    time.sleep(3)
    driver.quit()

def backForward():
    # 后退前进
    driver = webdriver.Chrome()
    url = "https://www.baidu.com"
    driver.get(url)
    driver.maximize_window()
    time.sleep(3)
    driver.get("https://www.taobao.com")
    time.sleep(3)
    driver.back()    # 跳转到前一页面
    time.sleep(3)
    driver.forward()  # 跳转到后一页面
    time.sleep(3)
    driver.quit()

if __name__ == "__main__":
    #noWindow()
    MaxWindow()
    #backForward()
```

```python
# 如何定位元素，查元素有很多种方法，比如id，name，class，xpath，css等
# 我们在浏览器上右键点击检查，就可以看到元素的属性

# 比如定位百度的输入框
def findElement():
    driver = webdriver.Chrome()
    url = "https://www.baidu.com"
    driver.get(url)
    driver.maximize_window()
    # 定位输入框，第一种方法，这里我们查ID
    driver.find_element_by_id("kw").send_keys("python")  # id定位并查询“python”
    
    time.sleep(2)
    # 定位输入框，第二种方法
    from selenium.webdriver.common.by import By
    driver.find_element(By.ID, "su").click()  # id定位并查询“python”，click()点击，clear()清空，send_keys()输入，还有键盘操作keys等等
    # continue_link = driver.find_element_by_partial_link_text('Conti')  根据部分文本定位，href中包含Conti的元素

    # 注意find_element_by_函数之间是可以嵌套的，因为有时候没办法只根据一个属性就定位到我们想要的元素位置

    # 假设一个页面中嵌套了多层窗口，那么我们就需要切换窗口，这时候就需要用到switch_to.frame()函数
    time.sleep(3)
    driver.quit()

```

```python
## 处理下拉框处理方式，网页中需要选择下拉框选项的场景
def downBlank():
    driver = webdriver.Chrome()
    url = "https://www.bilibili.com"
    driver.get(url)
    driver.maximize_window()
    
    # 通过xpath定位到下拉框
    first = driver.find_element_by_xpath("//input[@maxlength='100']").click()
    time.sleep(3)
    from selenium.webdriver.common.by import By
    # 这里很奇怪不能直接first.find_element，只能driver我也不懂
    ele = driver.find_element(By.XPATH, '//*[@id="i_cecream"]/div[2]/div[1]/div[1]/div/div/div/div/div[2]/div[1]/div[4]/div[2]')
    ele.click()
    time.sleep(5)
    
    driver.quit()

if __name__ == "__main__":
    
    downBlank()
```

```python
## 处理弹窗，我们经常遇到一些广告弹窗，验证弹窗等等，此时必须先关掉才能做其他事
def alert():
    driver = webdriver.Chrome()
    
    
    # 处理弹窗
    # 获取弹窗文本
    text = driver.switch_to.alert.text
    print(text)
    # 点击弹窗确认按钮
    driver.switch_to.alert.accept()
    # 点击弹窗取消按钮
    driver.switch_to.alert.dismiss()
    
    driver.quit()
```
