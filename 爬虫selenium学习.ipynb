```python

# -*- coding:utf-8 -*-
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
 
def noWindow():
    # 不开启浏览器得到返回值
    # 加载启动项，只开一个监听接口获取返回值
    option = webdriver.ChromeOptions()
    option.headless = True  # 不打开浏览器，后台运行
    
    # 通过url访问浏览器
    driver = webdriver.Chrome(chrome_options=option)
    url = "https://www.baidu.com"

    driver.get(url)
    
    time.sleep(3)
    driver.quit()

def MaxWindow():
    # 最大化浏览器
    driver = webdriver.Chrome()
    url = "https://www.baidu.com"
    driver.get(url)
    driver.maximize_window()
    
    time.sleep(3)
    driver.quit()

def backForward():
    # 后退前进
    driver = webdriver.Chrome()
    url = "https://www.baidu.com"
    driver.get(url)
    driver.maximize_window()
    time.sleep(3)
    driver.get("https://www.taobao.com")
    time.sleep(3)
    driver.back()    # 跳转到前一页面
    time.sleep(3)
    driver.forward()  # 跳转到后一页面
    time.sleep(3)
    driver.quit()

if __name__ == "__main__":
    #noWindow()
    MaxWindow()
    #backForward()
```

```python
# 使用js代码操作页面，比如滚动条上拉下拉
def js():
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get('https://search.jd.com/Search?keyword=imac&enc=utf-8&wq=imac')
    js = 'document.documentElement.scrollTop=10000'   # 滚动条拉到最下面获取当前页面所有的数据
    driver.execute_script(js)
    time.sleep(1)
    all = driver.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div/a/em')  # 找imac所有的商品名称
    print(len(all))
    for item in all:
        print(item.text)

    time.sleep(3)
    driver.quit()

if __name__ == "__main__":
    js()
```

```python
## selenium等待处理，浏览器自动停留来处理事件
from selenium.webdriver.support.wait import WebDriverWait
# 隐式等待
def implicitWait():
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get('https://search.jd.com/Search?keyword=imac&enc=utf-8&wq=imac')
    js = 'document.documentElement.scrollTop=10000'   # 滚动条拉到最下面获取当前页面所有的数据
    driver.execute_script(js)
    driver.implicitly_wait(5)  # 隐式等待，最长等5秒，如果5秒内加载完了就不等了
    all = driver.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div/a/em')  # 找imac所有的商品名称
    print(len(all))

    
    driver.quit()

# 显示等待
def driverWait():
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get('https://search.jd.com/Search?keyword=imac&enc=utf-8&wq=imac')
    js = 'document.documentElement.scrollTop=10000'   # 滚动条拉到最下面获取当前页面所有的数据
    driver.execute_script(js)
    WebDriverWait(driver, 5, 0.5).until(lambda elements:driver.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div/a/em'))  
    # 强制等待，指定浏览器、最长等5秒、时间间隔为0.5秒查询下是否找到，until指定一个响应函数如果函数执行成功了就不等待了
    all = driver.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div/a/em')  # 找imac所有的商品名称
    print(len(all))
    
    driver.quit()

# 这两种等待适用于静态的网页还好用一点，不然的话只能用强制等待sleep了
if __name__ == "__main__":
    # implicitWait()
    driverWait()
```
